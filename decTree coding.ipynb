```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/9_decision_tree/salaries.csv")
```

```python
data.head()
```

```python
X = data.drop('salary_more_then_100k', axis=1)
Y = data.salary_more_then_100k
```

```python
X
```

```python
Y
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
X['company_n'] = le_company.fit_transform(X['company'])
X['job_n']=le_job.fit_transform(X['job'])
X['degree_n'] = le_degree.fit_transform(X['degree'])
```

```python
X
```

```python
X_n = X.drop(['company', 'job', 'degree'],axis=1)
X_n
```

```python
from sklearn import tree
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train, X_test, Y_train, Y_test = train_test_split( X_n, Y, test_size=0.25, random_state=1)
```

```python
model = tree.DecisionTreeClassifier()
model.fit(X_train,Y_train)
```

```python
Y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(Y_test,Y_pred)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(Y_test, Y_pred)
```
